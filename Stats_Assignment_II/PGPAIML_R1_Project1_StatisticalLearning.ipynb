# 2. Print all the teams which have started playing between 1930-1980. Use Debut column. (Include both the years - 1930 and 1980) (5 points)

```python
##Replacing the Debut years which have the character '-' and considering only the start year as debut year.
## For example: 1929-30 considering the debut year as 1929

laliga_dataSet.replace( {'Debut':'\\d{4}-\\d{2}'},
                                {'Debut':laliga_dataSet['Debut'].str[0:4]},regex=True,inplace=True)
```

```python
## converting the debut object into int64, extracting the Teams and Debut year and sorting 
## them in Ascending order of Debut Year

print(laliga_dataSet[laliga_dataSet['Debut'].astype({'Debut':'int64'}) >= 1930]
      [laliga_dataSet['Debut'].astype({'Debut':'int64'}) <= 1980]
      [['Team','Debut']].sort_values(by='Debut',
        axis=0,
        ascending=True,
        inplace=False,
        kind='quicksort'))
```

# 4. Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points). 

```python
def Goal_diff_count(laliga_dataSet):
        goalDiffList = []
        maxMinGoalDiff_Dict ={}
        for i in range(len(laliga_dataSet)):
            maxMinGoalDiff_Dict.update({laliga_dataSet.loc[i,'Team']:
                                        int(laliga_dataSet.loc[i,'GoalsFor']) - int(laliga_dataSet.loc[i,'GoalsAgainst'])})
            goalDiffList.append("S.No: "+str(i)+" "+laliga_dataSet.loc[i,'Team']+" has a GoalsFor of about "
                                +str(laliga_dataSet.loc[i,'GoalsFor'])
                                + " and Goals Against them  "+str(laliga_dataSet.loc[i,'GoalsAgainst'])
                                +" So, the Goal Difference is "
                                +str(int(laliga_dataSet.loc[i,'GoalsFor']) - int(laliga_dataSet.loc[i,'GoalsAgainst'])))
        goalDiffList.append("-------------------------Team with Maximum Goal Difference--------------------")
        goalDiffList.append(str(max(maxMinGoalDiff_Dict, key =maxMinGoalDiff_Dict.get))
                            +" is the team has maximum goal difference of about "
                            +str(max(maxMinGoalDiff_Dict.values())))
        goalDiffList.append("-------------------------Team with Minimum Goal Difference--------------------")
        goalDiffList.append(str(min(maxMinGoalDiff_Dict, key =maxMinGoalDiff_Dict.get))
                            +" is the team has minimum goal difference of about "
                            +str(min(maxMinGoalDiff_Dict.values())))
        return goalDiffList
```

```python
for row in Goal_diff_count(laliga_dataSet):
        print(row)
```

# 5. Create a new column with the name Winning Percent and append it to the data set (5 points) Percentage of Winning = (GamesWon / GamesPlayed)*100 If there are any numerical error, replace it with 0% Print the top 5 teams which have the highest Winning percentage

```python
## Adding the new feature Winning Percent to the dataFrame using the above formula
laliga_dataSet['Winning Percent']= laliga_dataSet['GamesWon'].astype('int64')/laliga_dataSet['GamesPlayed'].astype('int64') *100
```

```python
## Replacing the numerical error using the fillna() Where the team Girona has the numerical error so the 
## Winning percent is 0.00
laliga_dataSet.fillna(0).tail(5)
```

```python
## Top 5 teams having the highest winning percent
print(laliga_dataSet[['Team','Winning Percent']].nlargest(5,'Winning Percent',keep="all"))
```

# 6.Group teams based on their Best position and print the sum of their points for all positions (7.5 points)

```python
laliga_best_position_groupBy = laliga_dataSet[['BestPosition','Team','Points']].astype({'Points':'int64'}).groupby(['BestPosition'])
```

```python
display(HTML(laliga_best_position_groupBy.aggregate({'Team':lambda x: " , ".join(x),'Points':'sum'}).to_html()))
```
